## 2) 지도객체에 마커 추가
일반 마커
아이콘 색상값 종류

'lightgreen', 'darkgreen', 'darkblue', 'cadetblue', 'orange', 'lightred', 'darkred', 'green', 'blue', 'black', 'lightblue', 'white', 'lightgray', 'red', 'pink', 'beige', 'gray', 'purple', 'darkpurple'

```python
# 새로운 지도 객체 생성
map_osm1 = folium.Map(location=[37.502552, 127.024866], zoom_start=17)

# 마커 객체 생성
# -> 마커가 표시될 위치 [위도,경도], 클리시 표시될 메시지, 아이콘 색상
marker1 = folium.Marker([37.502552, 127.024866], 
                        popup='이젠아카데미컴퓨터학원 강남', 
                        icon=folium.Icon(color='darkred'))

marker1.add_to(map_osm1)  # 마커 객체를 지도에 추가함
map_osm1                  # 지도 표시하기
```

#### 사용자 지정 아이콘, HTML 팝업

```python
# 새로운 지도 객체 생성
map_osm2 = folium.Map(location=[37.502552, 127.024866], zoom_start=17)

# HTML을 사용한 팝업
popup_html = folium.Popup("<font color='green' style='white-space: nowrap'><b>이젠아카데미컴퓨터학원</b></font>", parse_html=False)
#popup_html = folium.Popup("<div style='white-space: nowrap'><h3>이젠아카데미컴퓨터학원</h3><img src='https://magazine.hankyung.com/magazinedata/images/raw/201807/59de268d73bd099bda19c0659fc6bb7d.jpg' width='100%'><br/><p>tel: <a href='tel:02-532-6500'>02-532-6500</a></p><p>서울특별시 서초구 1303-37 서초W타워 13층</p></div>", parse_html=False)

# 사용자 지정 아이콘 이미지 사용
# --> 온라인 상의 URL, 내 컴퓨터 상의 상대,절대경로 모두 가능함
icon_img = folium.features.CustomIcon('http://itpaper.co.kr/data/ho.png', icon_size=(50, 50))

# 마커 객체 생성
marker2 = folium.Marker([37.502552, 127.024866], 
                        popup=popup_html,
                        icon=icon_img)

marker2.add_to(map_osm2) # 마커 객체를 지도에 추가함
map_osm2.save('map_osm2.html') #파일이 저장될 위치
map_osm2                 # 지도 표시하기
```

원형 마커(범위지정)

```python
# 새로운 지도 객체 생성
map_osm3 = folium.Map(location=[37.502552, 127.024866], zoom_start=17)

# 마커 객체 생성
marker1 = folium.Marker([37.502552, 127.024866],  
                        icon=folium.Icon(color='orange'))

# 원형마커 
marker3 = folium.CircleMarker([37.502552, 127.024866], 
                              radius=100,                 # 범위
                              color='#3186cc',            # 선 색상
                              fill_color='#3186cc')       # 면 색상

# 원형 마커 위에 아이콘 마커 올리기
marker3.add_to(map_osm3)
marker1.add_to(map_osm3)

map_osm3
```

## #03. 서울의 초등학교 위치
1) 데이터 수집

```python
# 엑셀 파일을 원격으로 내려받아 데이터프레임으로 변환 (다소 시간이 소요된다.)
xlsx = read_excel("http://itpaper.co.kr/data/school.xlsx", engine='openpyxl')
xlsx
```

```python
df = xlsx.filter(['학교명', '학교급구분', '소재지도로명주소', '위도', '경도'])
df
```

## 2) 서울시의 초등학교만 추출
1. `학교급구분` 필드 값이 초등학교
2. `소재지도로명주소`에 서울 이라는 단어가 포함된 경우
  
  LIKE 연산
  컬럼이름.str.contains('검색어')

```python
df2 = df.query("학교급구분 == '초등학교' and 소재지도로명주소.str.contains('서울특별시')")
df2
```

```python
# zoom_start: 배율 1~22 (여기서는 출력 안함)
map_osm = folium.Map(location=[37.566651, 126.978428], zoom_start=12)
map_osm
```

```python
for i in df2.index:
    # 행 우선 접근 방식으로 값 추출하기
    name = df2.loc[i, '학교명']
    lat = df2.loc[i, '위도']
    lng = df2.loc[i, '경도']
    
    # 추출한 정보를 지도에 표시
    marker = folium.Marker([lat,lng], popup=name)
    marker.add_to(map_osm)

# 저장과 지도 출력을 함께 사용할 경우 "저장 -> 출력"순으로 사용해야 함.
map_osm.save('서울의_초등학교_분포.html')
map_osm
```
