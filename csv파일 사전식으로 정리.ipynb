```python
dic = {} 

total = [0,0,0]


with open("traffic.csv", 'r', encoding = 'euc-kr') as f:
    line = f.readlines()
    for i, item in enumerate(line):
        if i > 0:
            tmp = item.strip().split(',')
            
            if tmp[0] + '년' not in dic:
                dic[tmp[0] + '년'] = {}
            
            dic[tmp[0] + '년'][tmp[1] + '월'] = {'발생건수' : tmp[2], '사망자수' : tmp[3], '부상자수' : tmp[4]}
    
    
    for i in dic:
        
        month_info = list(dic[i].values())
        total_year = [0,0,0]
        for index, j in enumerate(month_info):
           
            total_year[0] += int(j['발생건수'])
            total_year[1] += int(j['사망자수'])
            total_year[2] += int(j['부상자수'])
            
            print("%s %d월 교통사고 발생건수 %d" %(i, index+1, int(j['발생건수'])), end = '|')
            print("사망자수 %d" %int(j['사망자수']),end = '|')
            print("부상자수 %d" %int(j['부상자수']))
        
        total[0] += total_year[0]
        total[1] += total_year[1]
        total[2] += total_year[2]
        print('-'*100)
        print("%s 총 교통사고 발생건수 %d" %(i, total_year[0]), end = '|')
        print("%s 총 사망자수 %d" %(i, total_year[1]), end='|')
        print("%s 총 부상자수 %d" %(i, total_year[2]))
        print('-'*100)
        
    print('-'*100)
    print("총 교통사고 발생건수 %d" %(total[0]), end = '|')
    print("총 사망자수 %d" %(total[1]), end='|')
    print("총 부상자수 %d" %(total[2]))
    print('-'*100)
```

```python
import pandas as pd

dataframe = pd.DataFrame.from_dict({(i,j): dic[i][j]
                                    for i in dic.keys()
                                    for j in dic[i].keys()}, orient = 'index')

dataframe.발생건수 = pd.to_numeric(dataframe.발생건수)
dataframe.사망자수 = pd.to_numeric(dataframe.사망자수)
dataframe.부상자수 = pd.to_numeric(dataframe.부상자수)

fig, axes = plt.subplots(nrows=2, ncols=6)

dataframe.plot.bar(ax = axes[0,0])
dataframe.plot.bar(ax = axes[0,1])

```
