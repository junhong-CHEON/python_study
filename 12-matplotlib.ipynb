### wordcloud 색상처리

```python
# 글자 색상을 리턴 하기 위한 함수 정의하기
def make_colors(word, font_size, position, orientation, random_state, **kwargs):
    # case1) color = '#ff6600'

    # case2) rgb(255,255,255) 형식의 색상값 생성
    #random_state.seed()
    #r = random_state.randint(0,255)
    #g = random_state.randint(0,255)
    #b = random_state.randint(0,255)

    #color = 'rgb(%d, %d, %d)' %(r,g,b)

    # case3) hsl 형식 색상 지정
    # 색상(H) : 0 ~ 360 원형의 색상 띠
    # 채도(S) : 0 ~ 100%
    # 명도(L) : 0 ~ 100%
    random_state.seed()
    a = random_state.randint(30, 90)
    b = random_state.randint(30, 90)
    color = 'hsl(270,%d%%, %d%%)' % (a,b)
    return color


text = ''
with open('이상한나라의앨리스.txt', encoding='utf-8') as f:
    text = f.read()

wc = WordCloud(width = 2560, height= 1440, max_font_size=150,background_color='#ffffff')

wc.generate(text)

# 계산된 빈도수에 색상값을 정의.
# -> 색상값을 리턴해 줄 함수의 이름과 randaom_state 기능 사용 여부 설정
wc.recolor(color_func=make_colors, random_state=True)
wc.to_file('color.png')

```

## 한자어 처리

```python
# 한자는 다른 언어와 다르게 모든 글자가 개별적인 의미를 갖음
# 모든 글자를 원소로 갖는 리스트를 생성

# WordCloud 객체는 영어에 대해서만 빈도수 계산 가능
# 다른언어는 Counter 모듈 사용

from wordcloud import WordCloud
from collections import Counter


text = ''
with open('시경.txt', encoding='utf-8') as f:
    text = f.read()

tmp = list(text)
print(tmp)

# 빈 문자열, 줄바꿈, 쉼표 제거
hanja = []

ignore = [" ", '\n', ',', '.', '(', ')', '\U000f0703','\ufeff']

# 글자 리스트 처리

for item in tmp:
    if item not in ignore:
        hanja.append(item.strip())

count = Counter(hanja)

most = count.most_common(300)
print(most)
```

### 워드클라우드가 요구하는 형식으로 가공

```python
# 워드 클라우드는 {'단어' : '빈도수' } 형식의 딕셔너리를  요구한다.
tag = {}

for n,c in most: tag[n] = c
print(tag)

wc = WordCloud(font_path='batang', width = 2560, height= 1440, max_font_size=250,background_color='#ffffff',scale = 2.0)

wc.generate_from_frequencies(tag)

wc.to_file('시경.png')
```

## 형태소 분석

```python
from wordcloud import WordCloud
from collections import Counter
from konlpy.tag import Okt

text = ''
with open('대한민국헌법.txt', encoding = 'utf-8') as f:
    text = f.read()

nlp = Okt()

nouns = nlp.nouns(text)
print(nouns)

# 빈도수 계산
count = Counter(nouns)

most = count.most_common(100)
print(most)

# 딕셔너리로 가공

tags = {}

for n,c in most: tags[n] = c

wc = WordCloud(font_path='NanumGothic.ttf', width=2560, height=1440,scale=2.0,max_font_size=250)

wc.generate_from_frequencies(tags)

wc.to_file("대한민국헌법-주요단어.png")
print(tags)
```
