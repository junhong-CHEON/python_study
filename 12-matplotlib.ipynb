```python
import numpy as np

dic = {}

with open("grade.csv", 'r', encoding = 'euc-kr') as f:
    line = f.readlines()
    for i, item in enumerate(line):
        if i > 0:
            tmp = item.strip().split(',')
            arr = []
            for j in range(1, len(tmp)):
                arr.append(int(tmp[j]))
            dic[tmp[0]] = np.array(arr)
            
grade_average = []
keys = list(dic.keys())

for i in dic:
    grade_average.append(np.average(dic[i]))
    
pyplot.figure()

pyplot.plot(grade_average, label = 'average', color ='#ff00ff')

pyplot.legend()
pyplot.grid()
pyplot.title('Grade_average')
pyplot.xlabel('grade')
pyplot.ylabel('average')


# x축의 좌표에 대한 리스트 생성
xpos = list(range(0, len(grade_average)))
pyplot.xticks(xpos, keys)
pyplot.ylim(0,100)
pyplot.xlim(0,5)

pyplot.savefig('grade_average.png')
pyplot.show()
pyplot.close()
```

## 폰트설정

```python
# 윈도우와 맥을 구분해서 설정

import os
import sys
import matplotlib as mpl
from matplotlib import font_manager


# 시스템 글꼴 목록을 리스트로 저장
font_list = font_manager.findSystemFonts()
    
font_list.sort()
    
for file_path in font_list:
        
    # 폰트속성 객체 가져오기
    fp = font_manager.FontProperties(fname = file_path)
    try:
    # 폰트속성을 통해 파이썬에 설정해야하는 이름 조회
        font_name = fp.get_name()
    except:
        print("예외가 발생했습니다.")
        
    # 폰트경로와 폰트이름 출력
    print("%s >> %s" %(file_path, font_name))
```

```python
dic = {}

with open('covid19_now.csv', 'r', encoding = 'euc-kr') as f:
    lines = f.readlines()
    for i,line in enumerate(lines):
        tmp = line.strip().split(',')
        if i == 0:
            for j in range(0, len(tmp)):
                dic[tmp[j]] = []
            key_list = list(dic.keys())
        else:
            for j, v in enumerate(key_list):
                if j == 0:
                    dic[v].append(tmp[j]) # 도시이름 str
                else:
                    dic[v].append(int(tmp[j])) # 인원수 int


xpos = np.arange(len(dic['지역명']))
label = dic['지역명']

data_1 = np.array(dic['확진자']) - np.array(dic['전날확진자'])
data_2 = np.array(dic['치료중']) - np.array(dic['전날치료중'])

plt.figure()

# x축의 좌표값을 사용한 그래프 구현
plt.bar(xpos - 0.2, data_1, label = '오늘확진자', width=0.4, color='#f2e3f7')
plt.bar(xpos + 0.2, data_2, label = '오늘치료중', width=0.4, color='#ecc2f9')

plt.legend()
plt.grid()
plt.xlabel('도시')
plt.ylabel('인원수')
plt.title('도시별 오늘 확진자, 치료중 수')

plt.xticks(xpos,label)

for i ,v in enumerate(data_1):
    str_val = '%d명' %v
    plt.text(xpos[i] - 0.2, data_1[i], str_val, fontsize = 9, color = '#000000',
    horizontalalignment = 'center', verticalalignment = 'bottom')

for i ,v in enumerate(data_2):
    str_val = '%d명' %v
    if v >= 0:
        plt.text(xpos[i] + 0.2, data_2[i], str_val, fontsize = 9, color = '#000000',
        horizontalalignment = 'center', verticalalignment = 'bottom')
    else:
        plt.text(xpos[i] + 0.2, 0, str_val, fontsize = 9, color = '#000000',
        horizontalalignment = 'center', verticalalignment = 'bottom')
        

plt.savefig('answer.png', dpi = 200)
plt.show()
plt.close()
```

```python
from matplotlib import pyplot as plt 
import numpy as np

# 글꼴 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 12
plt.rcParams['figure.figsize'] = (12,8)

file_list = []

label = []
dic = {}

def graph_text(x, value, color, horizon, vertical):
    for i ,v in enumerate(x):
        str_val = '%d명' %value[i]
        plt.text(x[i], value[i], str_val, fontsize = 9, color = color,
        horizontalalignment = horizon, verticalalignment = vertical)

# 파일 가져오기
with open('website_visit.csv', 'r', encoding='euc-kr') as f:
    
    lines = f.readlines()
    
    for i, line in enumerate(lines):
        
        tmp = line.strip().split(',')

        if i == 0:
            label = tmp[1:] # 첫번째 행을 그래프의 라벨로 사용
        else:
            tmp_list = []
            for j, v in enumerate(tmp):
                if j == 0:
                    continue
                tmp_list.append(int(v)) # int 형식으로 바꿔서 임시 리스트에 저장
            dic[tmp[0]] = np.array(tmp_list) # 각 딕셔너리 키에 값 추가
                
    
# 그래프 설정
xpos = np.arange(len(label))

plt.figure()

color_set = ['green','orange','blue']

for i, v in enumerate(dic):
    plt.plot(dic[v], label = v, marker = 'o', linestyle = '-', color = color_set[i])

plt.legend()
plt.grid()
plt.xlabel('월')
plt.ylabel('방문객 수')
plt.ylim()
plt.xticks(xpos, label)


for i,v in enumerate(dic):
    graph_text(xpos, dic[v], color_set[i] , 'center', 'bottom')

file_list.append('visit_diff_year.png')
plt.savefig('visit_diff_year.png')
plt.show()
plt.close()
```

```python
average_list = []

key_list = list(dic.keys()) # label로 쓰기위한 키리스트
xpos = np.arange(len(key_list)) # 새로운 x 좌표 생성

for i in dic:
    average_list.append(np.average(dic[i]))
print(average_list)

plt.figure()

plt.bar(xpos, average_list, label = '웹사이트별 연평균 방문자', width=0.4, color=color_set)

plt.grid()

plt.xlabel('웹사이트')
plt.ylabel('평균 방문자 수')
plt.ylim(400,700)
plt.title('웹사이트별 연평균 방문자')
plt.xticks(xpos, key_list)

graph_text(xpos, average_list, 'black' , 'center', 'bottom') # 텍스트 표시

file_list.append('visit_average_year.png')
plt.savefig('visit_average_year.png')
plt.show()
plt.close()
```

```python
sum_list = []

for i in dic:
    sum_list.append(np.sum(dic[i])) # 각 1년간 총이용자 리스트 생성

ratio = sum_list
labels = key_list

explode = [0.0, 0.0, 0.0]

plt.figure()
plt.title('웹사이트 점유율')
plt.pie(ratio, labels = labels, colors = color_set, explode = explode,
    autopct = '%0.2f%%', shadow=False, startangle=90)

file_list.append('website_ratio.png')
plt.savefig('website_ratio.png')
plt.show()
plt.close()
```
