## 딕셔너리로부터 그래프 생성

```python
from sample import grade

keys = list(grade.keys())
values = list(grade.values())

pyplot.figure()

pyplot.plot(values, label = 'grade', color ='#ff00ff')

pyplot.legend()
pyplot.grid()
pyplot.title('Level Grade')
pyplot.xlabel('level')
pyplot.ylabel('point')
pyplot.savefig('line1-1.png')

# x축의 좌표에 대한 리스트 생성
xpos = list(range(0, len(values)))
pyplot.xticks(xpos, keys)
pyplot.ylim(0,100)
pyplot.xlim(0,3)

pyplot.savefig('line1-2.png')
pyplot.close()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import os
plt.rc('font',family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)

%config InlineBackend.figure_format = 'retina'
```

```python
import numpy as np

dic = {}

with open("grade.csv", 'r', encoding = 'euc-kr') as f:
    line = f.readlines()
    for i, item in enumerate(line):
        if i > 0:
            tmp = item.strip().split(',')
            dic[tmp[0]] = np.array([int(tmp[1]), int(tmp[2]), int(tmp[3]), int(tmp[4])])
        
grade_average = []
keys = list(dic.keys())

for i in dic:
    grade_average.append(np.average(dic[i]))
    
pyplot.figure()

pyplot.plot(grade_average, label = 'average', color ='#ff00ff')

pyplot.legend()
pyplot.grid()
pyplot.title('Grade_average')
pyplot.xlabel('grade')
pyplot.ylabel('average')
pyplot.savefig('line1-1.png')

# x축의 좌표에 대한 리스트 생성
xpos = list(range(0, len(grade_average)))
pyplot.xticks(xpos, keys)
pyplot.ylim(0,100)
pyplot.xlim(0,5)

pyplot.savefig('grade_average.png')
pyplot.show()
pyplot.close()
```
