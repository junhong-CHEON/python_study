```python
from pandas import *
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python

plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 16
plt.rcParams["figure.figsize"] = (20, 10)
plt.rcParams['axes.unicode_minus'] = False
```

# 3장 연습문제

## 문제1)

```python
dic = {'연령군' : ['10대', '20대' ,'30대', '40대'], '사람수' : [20,55,68,57], '%' :[10.0, 27.5, 34.0, 28.5]}

ans1_df = DataFrame(dic).set_index(['연령군'])
ans1_df


fig, ax = plt.subplots(1,2)
ans1_df.filter(['사람수']).plot.bar(rot = 0, ax = ax[0])
ax[0].set_title('사람수')
ax[0].grid()

ans1_df['%'].plot.pie(ax = ax[1], autopct='%0.1f%%')
ax[1].legend()
ax[1].set_title('컴퓨터 구입 고객의 연령별 사람비율')
ax[1].set_ylabel(None)
```

## 문제2)

```python
dic = {'항목': [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
      '청년층인구' : [9684, 9552, 9440, 9397, 9385, 9378, 9302, 9157, 9073, 8934],
      '취업자' : [3958, 3919, 3761, 3780, 3890, 3950, 4003, 3908, 3953, 3770],
      '실업자' : [313, 340, 300, 359, 397, 425, 408, 460, 437, 426]}

ans2_df = DataFrame(dic).set_index(['항목'])

fig, ax = plt.subplots(1,3)

ans2_df['청년층인구'].plot.bar(ax = ax[0], rot = 45)
ax[0].grid()
ax[0].set_title('청년층인구')
ax[0].set_ylim(8000,10000)

ans2_df['취업자'].plot(ax = ax[1], rot = 45)
ax[1].grid()
ax[1].set_title('취업자수')
ax[1].set_xticks(np.arange(min(ans2_df.index), max(ans2_df.index)+1))
ax[1].set_xticklabels(ans2_df.index, fontsize=15, color='black')

ans2_df['실업자'].plot(ax = ax[2], rot = 45, color = 'orange')
ax[2].grid()
ax[2].set_title('실업자수')
ax[2].set_xticks(np.arange(min(ans2_df.index), max(ans2_df.index)+1))
ax[2].set_xticklabels(ans2_df.index, fontsize=15, color='black')
```

```python
dic = {'연도': [1990, 1995, 2000, 2005, 2010, 2012, 2013, 2030, 2040],
      '총인구' : [42869, 45093, 47008, 48138, 49410, 50004, 50220, 52160, 51091],
      '인구성장률' : [0.99, 1.01, 0.84, 0.21, 0.46, 0.45, 0.43, 0.01, -0.39]}

ans3_df = DataFrame(dic).set_index(['연도'])

fig, ax1 = plt.subplots(1,1)


ans3_df['총인구'].plot.bar(color = 'skyblue',ax = ax1, rot = 0)
ax1.grid()
ax1.set_title('총인구')
ax1.set_ylabel('총인구', color = 'skyblue')
ax1.set_ylim(40000,55000)


ax2 = ax1.twinx()
ax2.plot(list(ans3_df['인구성장률']), color = 'pink', marker ='o')
ax2.set_ylabel('인구성장률', color = 'pink')

plt.show()
```
