```python
# csv데이터 각 리스트에 저장
with open("traffic.csv", "r" , encoding = 'euc-kr') as f:
    data_line = f.readlines()
    
    occur = []
    dead = []
    wound = []
    
    for i, line in enumerate(data_line):
        if i > 0:
            item = line.strip().split(",")
            
            occur.append(int(item[2]))
            dead.append(int(item[3]))
            wound.append(int(item[4]))
    print(occur)
    print(dead)
    print(wound)
```

# 1)각각의 데이터 범위

```python
print("한달 교통사고 발생량의 범위는 %d건 부터 %d건 까지이다. " %(min(occur), max(occur)))
print("한달 사망자수의 범위는 %d명 부터 %d명 까지이다. " %(min(dead), max(dead)))
print("한달 부상자수의 범위의 %d명 부터 %d명 까지이다. " %(min(wound), max(wound)))
```

# 2)평균 구하기

```python
occur_avg = sum(occur)/ len(occur)
dead_avg = sum(dead)/ len(dead)
wound_avg = sum(wound)/ len(wound)

print("한달 교통사고 발생의 평균 건수는 %0.2f건 이다. " %occur_avg)
print("한달 사망자의 평균은 수는 %0.2f명 이다. " %dead_avg)
print("한달 부상자의 평균은 수는 %0.2f명 이다. " %wound_avg)
```

# 3)분산 구하기

```python
def deviation(list, avg):
    dev_list = []
    for i in list:
        tmp = (avg - i) ** 2
        dev_list.append(tmp)
    
    return dev_list

occur_var = sum(deviation(occur, occur_avg))/len(occur)
dead_var = sum(deviation(dead, dead_avg))/len(dead)
wound_var = sum(deviation(wound, wound_avg))/len(wound)

print("한달 교통사고 발생의 분산은 %0.2f 이다. " %occur_var)
print("한달 사망자의 분산은 %0.2f 이다. " %dead_var)
print("한달 부상자의 분산은 %0.2f 이다. " %wound_var)
```

# 4)표준편차 구하기

```python
import math

occur_SD = math.sqrt(occur_var)
dead_SD = math.sqrt(dead_var)
wound_SD = math.sqrt(wound_var)

print("한달 교통사고 발생의 표준편차는 %0.2f 이다. " %occur_SD)
print("한달 사망자의 표준편차는 %0.2f 이다. " %dead_SD)
print("한달 부상자의 표준편차는 %0.2f 이다. " %wound_SD)
```
