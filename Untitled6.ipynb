```python
from pandas import DataFrame
from pandas import read_excel
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np

# 단순선형회귀 방법 (1)
from sklearn.linear_model import LinearRegression
# 단순선형회귀 방법 (2)
from statsmodels.formula.api import ols
```

```python
df = read_excel('http://itpaper.co.kr/data/cars.xlsx', engine='openpyxl')
df.head()
```

```python

plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 20

fig, ax = plt.subplots(1, 2, figsize=(20, 7))

for i, v in enumerate(df.columns):
    # numpy를 통한 히스토그램 추출
    # n -> 빈도수, bins -> 구간정보
    n, bins = np.histogram(df[v], bins=5, range=(0, df[v].max()))
    # 히스토그램 그리기
    sns.histplot(x=v, data=df, bins=5, binrange=(0, df[v].max()), ax=ax[i])
    
    ax[i].set_xticks(bins)
    ax[i].set_xticklabels(bins)
    ax[i].set_xlabel(v)

    # 출력할 데이터 수 만큼 반복
    for j, w in enumerate(n):    
        x = bins[j] + ((bins[j+1]-bins[j]) / 2)

        # 텍스트 출력 -> x좌표, y좌표, 내용, 글자크기, 색상, 텍스트가로정렬, 세로정렬
        ax[i].text(x, w, str(int(w)), fontsize=20, color='#ffffff', horizontalalignment='center', verticalalignment='top')
    
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (12, 6)

# 산점도 그래프
sns.scatterplot(x='speed', y='dist', data=df)
# 추세선 표시
sns.regplot(x='speed', y='dist', data=df)
plt.xlim(df['speed'].min()-1, df['speed'].max()+1)
plt.grid()
plt.show()
```

```python
# sklearn을 활용한 회귀분석의 경우 모든 값들이 개별적인 1차 배열에 담겨 있는 형태의 2차 배열로 전처리가 수행되어야 한다.
x = df['speed'].values.reshape(-1,1)
x
```

```python
y = df['dist'].values
y
```

```python
fit = LinearRegression().fit(x, y)
fit
```

```python
a = fit.coef_
b = fit.intercept_
print('f(x) = %fx + %f' %(a,b))
```

```python
def f(x):
    global a,b
    return a * x  + b
```

```python
result1 = f(x)
result1
```

```python
result2 = fit.predict(x)
result2
```

```python
residual = y - result2
residual
```

```python
DataFrame(residual).describe()
```

```python
fit.score(x,y)
```

```python
import scipy.stats as stats

stats.probplot(residual, plot=plt)
```

```python
res = ols('dist ~ speed', data=df).fit()
res.summary()
```

[회귀분석의 결과 해석](https://www.notion.so/d0b6be933cb0434db1913867e6059e84)

```python
# 잔차
res_df = DataFrame(res.resid)
res_df.head()
```

```python
res_df.describe()
```

```python
a = res.params.speed
b = res.params.Intercept
print('y = %fx + %f' % (a,b))
```

```python
res.fittedvalues
```

```python
res.predict({'speed' : list(range(10,5))})
```

```python
plt.rcParams["figure.figsize"] = (12, 6)
plt.rcParams["font.size"] = 16

sns.scatterplot(x='speed', y='dist', data=df)

# 학습에 사용된 반응변수(x)에 따른 설명변수(y)의 추정치를 활용하여 직선 추가
sns.lineplot(x='speed', y=res.fittedvalues, data=df, color='red')

plt.xlim(df['speed'].min()-1, df['speed'].max()+1)
plt.grid()
plt.show()
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["figure.figsize"] = (12, 6)
plt.rcParams["font.size"] = 16
plt.rcParams['axes.unicode_minus'] = False

# 예측값
fitted = res.predict(df)
# 잔차(실데이터-예측값)
residual = df['dist'] - fitted

sns.regplot(x=fitted, y=residual, lowess=True, line_kws={'color': 'red'})
plt.plot([fitted.min(), fitted.max()], [0, 0], '--', color='grey')
plt.grid()
plt.show()
```
