```python
# ChromeDriver 자동 설치 모듈
import chromedriver_autoinstaller
# Chrome을 제어하기 위한 객체
from selenium import webdriver
# Chrome이 웹 페이지 로딩을 완료 할 때까지 최대 n초간 대기하는 기능.
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.alert import Alert
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
from pandas import DataFrame
# 파이썬 프로그램에 지정된 시간동안 랙을 거는 기능을 위해 사용
import time

import itertools
import openpyxl
import numpy as np
```

```python
# 크롬드라이버 자동 설치
chromedriver_autoinstaller.install()
```

```python
df = DataFrame()
```

```python
# 크롬드라이버를 통해 크롬을 실행시킴
# -> driver 객체는 Chrome 자체

driver = webdriver.Chrome()
# 크롬브라우저가 준비될 때 까지 최대 5초씩 대기

driver.implicitly_wait(5)

driver.get("https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=100")
    # 페이지를 이동하는 동안 최대 2초간 대기
time.sleep(2)
```

```python

#lnb > ul > li:nth-child(4) > a
#lnb > ul > li:nth-child(5) > a

document = []
label = []
labels = ['정치', '경제', '사회' ,'생활/문화', '세계', 'IT/과학']
for thema_num in range(4,10):
    
    page = 1
    next = 1
    
    label_val = labels[thema_num - 4]
    while(True):
        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")
        title = soup.select('#section_body ul')

        for i in title:
            li_list = i.select('li')
            for j in li_list:
                if bool(j.select('dl > .photo')):
                    a = j.select('dl > dt:nth-of-type(2) > a')
                else:
                    a = j.select('dl > dt:nth-of-type(1) > a')
                document.append(a[0].text.strip())
                label.append(label_val)
    
        next += 1
        if page == 1 and next == 12:
            print('[%s/%d쪽] -----> 크롤링 완료' %(label_val,page*10))
            page += 1
            next = 3
        elif next == 13:
            print('[%s/%d쪽] -----> 크롤링 완료' %(label_val,page*10))
            page += 1
            next = 3
        try:
            next_page = WebDriverWait(driver, 3).until(lambda x: x.find_element_by_css_selector("#paging > a:nth-child(%s)" %next))
            next_page.click()
            time.sleep(1)
        except:
            break
    next_thema = WebDriverWait(driver, 3).until(lambda x: x.find_element_by_css_selector("#lnb > ul > li:nth-child(%s) > a" %thema_num))
    next_thema.click()
    time.sleep(1)
```

```python
df = DataFrame({'document' : document, 'label' : label})
```

```python
df = df.drop_duplicates(['document','label'], keep='last').reset_index()
df = df.drop(['index'], axis = 1)
```
