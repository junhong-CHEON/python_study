```python
import requests
from bs4 import BeautifulSoup

user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.120 Safari/537.36"
header_info = {'User-agent': user_agent, 'referer': None}
header_info

content_url = "https://arca.live/b/genshin/26806010"

r = requests.get(content_url, headers = header_info)

if r.status_code != 200:
    # ì—ëŸ¬ì½”ë“œì™€ ì—ëŸ¬ë©”ì‹œì§€ë¥¼ ë¬¸ìì—´ë¡œ êµ¬ì„±
    err_msg = "%d %s ì—ëŸ¬ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤." % (r.status_code, r.reason)
    # ê°•ì œë¡œ ì—ëŸ¬ë¥¼ ë°œìƒì‹œí‚¨ë‹¤.
    raise Exception(err_msg)
    
r.encoding="utf-8"
html = r.text
#html
```

```python
# ì›¹ í˜ì´ì§€ì˜ ì†ŒìŠ¤ì½”ë“œë¥¼ ì‚¬ìš©í•˜ì—¬ HTML ë¶„ì„ ê°ì²´ ìƒì„±
soup = BeautifulSoup(html, 'html.parser')

# CSS ì„ íƒìë¥¼ í™œìš©í•˜ì—¬ ê°€ì ¸ì˜¤ê¸°ë¥¼ ì›í•˜ëŠ” ë¶€ë¶„ ì§€ì •
# -> ì—¬ëŸ¬ ê°œì˜ ì„ íƒìë¥¼ ì‚¬ìš©í•´ì•¼ í•  ê²½ìš° ì½¤ë§ˆ(,)ë¡œ êµ¬ë¶„
#
#
selector = soup.select('.article-content video')
#
#

# ê²°ê³¼ ê²€ì‚¬
if len(selector) == 0:
    # ì—ëŸ¬ì½”ë“œì™€ ì—ëŸ¬ë©”ì‹œì§€ë¥¼ ë¬¸ìì—´ë¡œ êµ¬ì„±
    err_msg = "ì§€ì •ëœ ì„ íƒìì— ëŒ€í•´ì„œ ì¸ì‹ëœ êµ¬ì¡°ê°€ ì—†ìŠµë‹ˆë‹¤."
    # ê°•ì œë¡œ ì—ëŸ¬ë¥¼ ë°œìƒì‹œí‚¨ë‹¤.
    raise Exception(err_msg)

print(len(selector))
print("-" * 50)
selector
```

```python
# ë‰´ìŠ¤ê¸°ì‚¬ì˜ ë³¸ë¬¸ URLì„ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
url_list = []

# ë¦¬ìŠ¤íŠ¸ì˜ ì›ì†Œë“¤ì— ëŒ€í•œ ë°˜ë³µ ì²˜ë¦¬
for item in selector:
    # ê° ì›ì†Œ(ë§í¬)ì— ì†ì„±ë“¤(attrs) ì¤‘ì— href ì†ì„±ì´ ìˆë‹¤ë©´ ê·¸ ê°’ì„ ë³„ë„ë¡œ ì¤€ë¹„í•œ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
    if "src" in item.attrs:
        url_list.append('https:' + item['src'])

url_list
url = url_list[0]
```

```python
r = requests.get(url, headers=header_info)

totalbits = 0
if r.status_code == 200:
    with open('asd.mp4', 'wb') as f:
        for chunk in r.iter_content(chunk_size=1024):
            if chunk:
                totalbits += 1024
                print("Downloaded",totalbits*1025,"KB...")
                f.write(chunk)
```
