- 표본의 크기가 상이하기 때문에 봉페르니교정 방법을 사용하였다.
- 볼베어링를 고정으로 했을때 대부분 진동과 소음의 관계는 유의미 하였다.
- 진동를 고정으로 했을때 볼베어링와 소음간의 관계는 유의미 하지 않았다.
- 이를 보아 볼베어링과 진동간의 상호작용 관계에서 진동 요인이 미치는 영향이 더 크다고 볼 수 있다.
