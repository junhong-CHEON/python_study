# 문제1)

```python
from pandas import DataFrame
from pandas import read_excel

iris = read_excel('http://itpaper.co.kr/data/iris.xlsx',engine='openpyxl')


iris_mean = DataFrame(iris.describe().mean(), columns = ['mean'])

iris_mean.to_excel('iris_mean.xlsx', engine = 'openpyxl')
```

# 문제3)

```python
from matplotlib import pyplot as plt
import seaborn as sns

남자 = [49, 86, 40, 45, 48, 93, 97, 58, 58, 98, 58, 82, 52, 56, 50, 85, 80, 60, 62, 80, 62, 72, 65, 60, 64, 70, 78, 67, 69, 88]
여자 = [60, 72, 66, 65, 75, 78, 62, 64, 74, 58, 68, 72, 67, 61, 62, 72, 79, 71, 74, 73]

남자수 = ['남자'] * len(남자)
여자수 = ['여자'] * len(여자)

점수 = 남자 + 여자
성별 = 남자수 + 여자수

df = DataFrame(index = range(0,len(점수)), columns = ['성별','점수'])
df['성별'] = 성별
df['점수'] = 점수
df
```

```python
남자점수 = df.query("성별 == '남자'")
여자점수 = df.query("성별 == '여자'")

plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 14
plt.rcParams["figure.figsize"] = (15, 10)

step = 5

fig, ax = plt.subplots(1,2, figsize= (15,5))
sns.histplot(data=남자점수, x = '점수', ax=ax[0], bins = step)
sns.histplot(data=여자점수, x = '점수', ax=ax[1], bins = step)
```
