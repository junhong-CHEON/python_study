```python
from pandas import DataFrame
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
from scipy import stats
from statsmodels.formula.api import ols
import scipy
```

```python
df = DataFrame({'월' : ['1월', '2월', '3월', '4월', '5월', '6월' ,'7월', '8월', '9월' ,'10월', '11월', '12월'],
               '광고비' : [62, 142, 122, 130, 185, 121, 101, 199, 221, 139, 169, 237],
               '매출액' : [100, 299, 298, 390, 590, 200, 190, 460, 660, 350, 650, 880]}).set_index('월')

df
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 16
plt.rcParams["figure.figsize"] = (12, 6)
plt.rcParams["axes.unicode_minus"] = False
```

## 1)

```python
fig, ax = plt.subplots(2,2, figsize = (20,10))
fig.tight_layout(pad=4.0)

plt.suptitle('광고비, 매출액 hist,box', fontsize = 30, color = 'blue')

ax_ind = 0
for i in df:
    if i == '광고비':
        n, bins = np.histogram(df[i], bins = 5, range = (0, 250))
        df[i].hist(ax = ax[0][ax_ind], bins = 5, range = (0, 250),rwidth = 0.8)
        x_label = []
        for j in bins:
            x_label.append('%d백만원' %j)
    else:
        n, bins = np.histogram(df[i], bins = 9, range = (0, 900))
        df[i].hist(ax = ax[0][ax_ind], bins = 9, range = (0, 900),rwidth = 0.8, color = 'green')
        x_label = []
        for j in bins:
            x_label.append('%d억원' %j)
            
    ax[0][ax_ind].set_title('%shist' %i, fontsize = 20, color = 'red')
    ax[0][ax_ind].set_xticks(bins)
    ax[0][ax_ind].set_xticklabels(x_label,rotation=15)
    ax[0][ax_ind].set_xlabel('%s' %i)
    
    for i,v in enumerate(n):
        txt = '%d' %v
        ax[0][ax_ind].text(bins[i] + (bins[i+1] - bins[i])/2, v, txt,
                          horizontalalignment = 'center', verticalalignment = 'bottom')
    ax_ind += 1
ax_ind = 0

for i in df:
    df.boxplot(i, ax = ax[1][ax_ind])
    ax[1][ax_ind].set_title('%sBoxPlot' %i, fontsize = 20, color = 'red')
    ax_ind += 1
    
```

## 2)
### 귀무가설
- 광고비와 매출액간의 평균에는 차이가 없다.
### 대립가설
- 광고비와 매출액간의 평균에는 차이가 있다.

```python
normal1 = stats.shapiro(df['광고비'])
normal2 = stats.shapiro(df['매출액'])

print(normal1)
print(normal2)
```

- pvalue > 0.05 이므로 정규성을 만족한다.

```python
levene = stats.levene(df['광고비'],df['매출액'])

print(levene)
```

- pvalue < 0.05 이므로 등분산성을 만족하지 못한다. 따라서 두 표본은 가설을 입증하기에 다소 부적합하다.

```python
tTestResult = stats.ttest_ind(df['광고비'], df['매출액'])
tTestResult
```

- t-test의 결과로 pvalue < 0.05 이기떄문에 5프로 유의수준에서 귀무가설을 기각할 수 있다.
- 하지만 두 표본이 등분산성을 가지지 않기 때문에 신뢰도가 떨어질 수 있다.

## 3)

```python
np.cov(df['광고비'], df['매출액'])
```

- 결과가 양수값 이므로 비례관계

```python
상관계수, pvalue = stats.pearsonr(df['광고비'] , df['매출액'])
print('상관계수 : %f, pvalue: %f' %(상관계수, pvalue))
```

- 상관계수가 1에 가까워 양의 상관관계를 가지고있다.
- pvalue < 0.05 이기때문에 대립가설을 채택하여 상관분석이 가능하다.

## 4)

```python
res = ols('매출액 ~ 광고비', data = df).fit()
res.summary()
```

- R-squared(결정계수) 값이 0.855로 선형성을 가지고 있다고 볼 수 있다.
- 판매액에 대하여 광고비로 예측하는 회귀분석을 실시한 결과, 이 회귀모형은 통계적으로 유의하다<br>
`F(1,10) = 58.74, p < 0.05`
- 광고비의 회귀계수는 4.1833으로, 매출액에 대하여 유의미한 예측변인인 것으로 타나남.<br>
`t(10) = 7.664, p < 0.05`
- y 절편값은 -215.0029 이고, 따라서 회귀식은 다음과 같음

```python
a = res.params.광고비
b = res.params.Intercept
print('y = %fx + (%f)' % (a,b))
```

## 5)

```python
fitted = res.predict(df)
residual = df['매출액'] - fitted

sns.regplot(x = fitted, y = residual, lowess=True, line_kws={'color' : 'red'}, label = '잔차')
plt.plot([fitted.min(), fitted.max()], [0, 0], '--', color = 'grey')
plt.title('선형성 확인')
plt.legend()
plt.grid()
plt.show()
```

- 예측값에 따른 잔차의 추세가 0과 가까운 범위에 있으므로 잔차가 일정하게 분포되어 있다.

```python
sr = scipy.stats.zscore(residual)
(x, y), fig = scipy.stats.probplot(sr)
sns.scatterplot(x=x, y=y)
plt.plot([-3, 3], [-3, 3], linestyle='--', color='grey')
plt.title('잔차 정규성 검사')
plt.grid()

plt.show()
```

- Q-Q plot의 결과로 잔차의 정규성을 확인할 수 있다.

```python
stats.shapiro(residual)
```

- 유의수준 5%에서 잔차의 정규성을 충족한다.

```python
sns.regplot(x=fitted, y=np.sqrt(np.abs(sr)), lowess=True, line_kws={'color': 'red'}, label = '잔차')
plt.grid()
plt.legend()
plt.title('잔차 등분산성 확인')
plt.show()
```

- 잔차가 일정 범위 안에서 특정한 패턴을 가지지 않기 때문에 등분산성을 만족한다.

```python
목표매출액 = np.linspace(900, 950)

def 추정광고비(매출액):
    a = res.params.광고비
    b = res.params.Intercept
    return (매출액 - b)/a

추정광고비_list = []

plt.figure(figsize = (50,10))
for i in 목표매출액:
    추정광고비_list.append(추정광고비(i))
    
p1 = plt.bar(추정광고비_list, 목표매출액, width = 0.2, color = 'orange')
plt.xticks(추정광고비_list, rotation = 15)
plt.bar_label(p1,fontsize = 13)
plt.title('광고비에 따른 목표 매출액 예상')
plt.xlabel('추정 광고비(백만원)')
plt.ylabel('목표 매출액(억원)')
plt.ylim(890, 960)
plt.grid()


```
